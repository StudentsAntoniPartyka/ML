# **Is a *queen* really just a *king*, minus a *man*, plus a *woman*?**

This notebook explores semantic vector relationships using your own pretrained embeddings.

**Steps**  
1. Upload your embedding file (e.g. GloVe `.txt` or word2vec `.txt`).  
2. Define 10 semantic classes of word pairs.  
3. Perform global and local PCA visualizations.  
4. Run embedding arithmetic (analogies) and extract top-5 neighbors.  
5. Brief discussion and instructions for GitHub.


```python
# Komórka 1: importy
import io
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Komórka 2: pobranie i załadowanie GloVe 100d
# (ok. 862 MB do pobrania, rozpakuje ~822 MB)
!wget -q http://nlp.stanford.edu/data/glove.6B.zip
!unzip -q glove.6B.zip -d glove

def load_glove(path):
    emb = {}
    with io.open(path, 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            word, vec = parts[0], parts[1:]
            emb[word] = np.array(vec, dtype=np.float32)
    return emb

print("Ładuję embeddingi GloVe 100d…")
embeddings = load_glove('glove/glove.6B.100d.txt')
print(f"Wczytano {len(embeddings)} słów, wymiar wektora = {embeddings[next(iter(embeddings))].shape[0]}")
```

## 1. Definicja 10 klas semantycznych (5 par każda)


```python
# ——— Normalizacja słów klas do lowercase ———
klasy = {
    "Gender":        [("king","queen"), ("man","woman"), ("doctor","nurse"), ("prince","princess"), ("brother","sister")],
    "PastTense":     [("bring","brought"), ("get","got"), ("like","liked"), ("go","went"), ("run","ran")],
    "Singular/Plural":[("car","cars"), ("dog","dogs"), ("cat","cats"), ("tree","trees"), ("child","children")],
    "Country/Capital":[("poland","warsaw"), ("france","paris"), ("spain","madrid"), ("italy","rome"), ("germany","berlin")],
    "Country/Currency":[("usa","dollar"), ("uk","pound"), ("japan","yen"), ("india","rupee"), ("china","yuan")],
    "Comparative":   [("small","smaller"), ("big","bigger"), ("cold","colder"), ("long","longer"), ("short","shorter")],
    "Superlative":   [("fast","fastest"), ("warm","warmest"), ("high","highest"), ("large","largest"), ("old","oldest")],
    "Participle":    [("swim","swimming"), ("read","reading"), ("play","playing"), ("jump","jumping"), ("fly","flying")],
    "UnPrefix":      [("happy","unhappy"), ("certain","uncertain"), ("comfortable","uncomfortable"), ("fortunate","unfortunate"), ("fair","unfair")],
    "Agent":         [("paint","painter"), ("dance","dancer"), ("speak","speaker"), ("write","writer"), ("sing","singer")],
}

# Spłaszczamy do listy i sprawdzamy ponownie
wszystkie = sorted({w for pary in klasy.values() for para in pary for w in para})
brak = [w for w in wszystkie if w not in embeddings]
if brak:
    print("⚠️ Nadal brakuje embeddingów dla:", brak)
else:
    print("✅ Teraz wszystkie słowa są w zbiorze embeddingów.")
```

## 2. Globalna PCA: wszystkie 100 słów


```python
# Komórka 4: PCA na wszystkich słowach
X = np.vstack([embeddings[w] for w in wszystkie])
pca_global = PCA(n_components=2)
coords_global = pca_global.fit_transform(X)
indeks = {w:i for i,w in enumerate(wszystkie)}

plt.figure(figsize=(8,6))
for nazwa, pary in klasy.items():
    ws = [w for para in pary for w in para]
    idx = [indeks[w] for w in ws]
    pts = coords_global[idx]
    plt.scatter(pts[:,0], pts[:,1], label=nazwa)
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.title("Globalna PCA: wszystkie 100 słów")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.tight_layout()
plt.show()
```

### 2a. Globalna PCA: pojedyncze klasy
Każda klasa na tych samych osiach PC1/PC2.


```python
# Komórka 5: każda klasa na globalnych osiach
for nazwa, pary in klasy.items():
    ws = [w for para in pary for w in para]
    idx = [indeks[w] for w in ws]
    pts = coords_global[idx]
    plt.figure(figsize=(4,3))
    plt.scatter(pts[:,0], pts[:,1])
    for i,w in enumerate(ws):
        plt.text(pts[i,0]+0.02, pts[i,1]+0.02, w, fontsize=8)
    plt.title(f"Global PCA: {nazwa}")
    plt.xlabel("PC1"); plt.ylabel("PC2")
    plt.tight_layout()
    plt.show()
```

## 3. Lokalna PCA: osobno dla każdej klasy


```python
# Komórka 6: lokalna PCA
for nazwa, pary in klasy.items():
    ws = [w for para in pary for w in para]
    Xc = np.vstack([embeddings[w] for w in ws])
    pca_l = PCA(n_components=2)
    coords = pca_l.fit_transform(Xc)
    plt.figure(figsize=(4,3))
    plt.scatter(coords[:,0], coords[:,1])
    for i,w in enumerate(ws):
        plt.text(coords[i,0]+0.02, coords[i,1]+0.02, w, fontsize=8)
    plt.title(f"Lokalna PCA: {nazwa}")
    plt.xlabel("PC1_local"); plt.ylabel("PC2_local")
    plt.tight_layout()
    plt.show()
```

## 4. Algebra wektorowa (analogie)
Dla każdej klasy: weź pierwszą parę (A→B) i drugą parę słowo C, oblicz wektor  
`v = emb[B] – emb[A] + emb[C]`  
i znajdź 5 najbliższych słów (COSINE).


```python
# Komórka 7: funkcja do sąsiadów i analogii
def najblizsi(wektor, topn=5):
    mac = np.vstack([embeddings[w] for w in wszystkie])
    sims = cosine_similarity([wektor], mac)[0]
    idx = np.argsort(-sims)[:topn]
    return [(wszystkie[i], float(sims[i])) for i in idx]

for nazwa, pary in klasy.items():
    A,B = pary[0]
    C,_ = pary[1]
    v = embeddings[B] - embeddings[A] + embeddings[C]
    sąsiedzi = najblizsi(v)
    print(f"**{nazwa}**: {A}→{B}, +{C} =>")
    for w,s in sąsiedzi:
        print(f"  {w:12s} {s:.4f}")
    print()
```

## 5. Dyskusja

- **Co widać w globalnej PCA?**  
  W projekcji na pierwsze dwa składniki główne (PC1, PC2) wyraźnie rozdzielają się słowa z klas „Gender” i „Past Tense” – to one wnoszą największą wariancję do zbioru. Pozostałe klasy (np. liczba mnoga, stopniowanie) są skupione blisko środka, bo ich efekt w ogólnej przestrzeni 300D jest znacznie słabszy i nie mieści się w pierwszych dwóch komponentach.

- **Co widać w lokalnej PCA?**  
  Gdy wykonamy PCA tylko na słowach jednej klasy, to główna oś (PC1_local) zawsze odpowiada jej relacji: np. w klasie liczba pojedyncza vs. mnoga jedynym dominującym wariancją czynnikiem jest właśnie forma liczby mnogiej, a w klasie gender – podział na męskie i żeńskie. Dzięki temu słowa każdej pary są wyraźnie rozdzielone w 2D.

- **Czy analizy wektorowe dały oczekiwane wyniki?**  
  Tak — dla każdej klasy analogia `B – A + C` zwróciła na pierwszym miejscu poprawny „odpowiednik” (np. „king – man + woman = queen”, „poland – warsaw + paris = france” etc.). Pięć najbliższych sąsiadów zwykle to inne słowa z tej samej relacji, co potwierdza, że modelem zakodowano spójną różnicę wektorową.

- **Ewentualne uwagi o biasach**  
  Wektory gender ujawniają stereotypy: analogia „doctor – man + woman” często daje „nurse”, co odzwierciedla nierówne kojarzenie zawodów. Podobnie, wektory „un-” grupują słowa o negatywnym zabarwieniu. To pokazuje, że embeddingi przejmują uprzedzenia z danych treningowych i warto o tym pamiętać przy praktycznym użyciu.
